```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd
```

```python
data=pd.read_csv('https://raw.githubusercontent.com/ezioauditore-tech/AI/main/datasets/Position_Salaries.csv')
data
```

```python
x=data.iloc[:,1:2].values
x
```

```python
y=data.iloc[:,2].values
y
```

```python
from sklearn.linear_model import LinearRegression
lin_reg=LinearRegression()
lin_reg.fit(x,y)
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly_reg2=PolynomialFeatures(degree=2)
x_poly=poly_reg2.fit_transform(x)
lin_reg2=LinearRegression()
print(x_poly)
lin_reg2.fit(x_poly,y)
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly_reg3=PolynomialFeatures(degree=3)
x_poly3=poly_reg3.fit_transform(x)
lin_reg3=LinearRegression()
print(x_poly3)
lin_reg3.fit(x_poly3,y)
```

# LINEAR REGRESSION

```python
plt.scatter(x,y,color='red')
plt.plot(x,lin_reg.predict(x),color='blue')
plt.xlabel('Position level')
plt.ylabel('salary')
plt.show()
```

# POLYNOMIAL REGRESSION WITH DEGREE 2 AND 3

```python
plt.scatter(x,y,color='red')
plt.plot(x,lin_reg2.predict(x_poly),color='blue')
plt.plot(x,lin_reg3.predict(x_poly3),color='green')
plt.xlabel("postion level")
plt.ylabel("salary")
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
df=pd.read_csv('https://raw.githubusercontent.com/ezioauditore-tech/AI/main/datasets/Real%20estate.csv')
df
```

```python
x=df.drop("Y house price of unit area",axis=1)
x
```

```python
y=df['Y house price of unit area']
y
```

```python
polynomial_reg=PolynomialFeatures(degree=2,include_bias=False)
```

```python
poly_feat=polynomial_reg.fit_transform(x)
poly_feat.shape
```

```python
x_train,x_test,y_train,y_test=train_test_split(poly_feat,y,test_size=0.3,random_state=101)
```

```python
polymodel=LinearRegression()
```

```python
polymodel.fit(x_train,y_train)
```

```python
y_pred=polymodel.predict(x_test)
```

```python
pd.DataFrame({'y_test':y_test,'Y_pred':y_pred,'Residuals':(y_test-y_pred)})
```

```python
poly_MAE=metrics.mean_absolute_error(y_test,y_pred)
poly_MSE=metrics.mean_squared_error(y_test,y_pred)
poly_RME=np.sqrt(poly_MSE)
print(poly_MAE)
print(poly_MSE)
print(poly_RME)
```

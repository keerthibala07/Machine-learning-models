```python
x_train=iris.drop('Species',axis=1)
```

```python
y_train=iris['Species']
y_train
```

```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import pandas as pd
```

```python
from sklearn.model_selection import train_test_split
x_train1,x_test1,y_train1,y_test1=train_test_split(x_train,y_train,random_state=0)
```

```python
weak_learner = DecisionTreeClassifier(max_depth=2)
```

```python
adaboost_classifier = AdaBoostClassifier(base_estimator=weak_learner, n_estimators=500, random_state=42)
```

```python
adaboost_classifier.fit(x_train1, y_train1)

# Make predictions on the test data
y_pred = adaboost_classifier.predict(x_test1)

# Evaluate the accuracy
accuracy = accuracy_score(y_test1, y_pred)
print(f'Accuracy: {accuracy:.2f}')
```

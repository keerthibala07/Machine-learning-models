```python
drop=['Name','Age',"SibSp",'Ticket','Cabin','Parch','Embarked']
train_test.drop(labels=drop,axis=1,inplace=True)
```

```python
train_tes_d=pd.get_dummies(train_test,columns=['Sex'])
```

```python
train_tes_d.fillna(value=0.0,inplace=True)
```

```python
x1_train=train_tes_d.values[0:891]
x1_test=train_tes_d.values[891:]

```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
x_train_scale=scaler.fit_transform(x1_train)
x_test_scale=scaler.transform(x1_test)#already fitted no need to give fit
```

```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import pandas as pd
```

```python
from sklearn.model_selection import train_test_split
x_train1,x_test1,y_train1,y_test1=train_test_split(x_train_scale,y_train,random_state=0)
```

```python
weak_learner = DecisionTreeClassifier(max_depth=1)
```

```python
adaboost_classifier = AdaBoostClassifier(estimator=weak_learner, n_estimators=800, random_state=42)
```

```python
adaboost_classifier.fit(x_train1, y_train1)

# Make predictions on the test data
y_pred = adaboost_classifier.predict(x_test1)

# Evaluate the accuracy
accuracy = accuracy_score(y_test1, y_pred)
print(f'Accuracy: {accuracy:.2f}')
```
